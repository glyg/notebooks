```python
scalp = io.imread('image0012.jpg')
scalp2 = io.imread('image0011.jpg')
```

```python
from skimage import feature
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(feature.canny(gray_scalp2))
```
