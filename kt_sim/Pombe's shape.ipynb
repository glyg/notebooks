# Pombe's shape

We're looking for an equation fitting the shape of the fission yeast, s.Pombe

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

## Answer 1: a simple ellipse

The good point here is that we have a pretty simple formula:
$$
\begin{eqnarray}
x &= \ell \cos(\theta) / 2 \\
y &= w \sin(\theta) / 2
\end{eqnarray}
$$

Where $\ell$ is the yeast length along its long axis and $w$ its width.

```python
def cortex1(length, width, num_points=100):
    thetas = np.linspace(0, 2*np.pi, num_points)
    xs = length/2. * np.cos(thetas)
    ys = width/2. * np.sin(thetas)
    return xs, ys
```

```python
fig, ax = plt.subplots()
xs, ys = cortex1(14, 4)
ax.plot(xs, ys, 'k-', lw=2)
ax.set_aspect('equal')
ax.set_ylim(-2.5, 2.5);
```

## Answer 2: a flatten ellipsis (kind of)

We can set a more complicated shape, still with a single equation, that will give something looking more like a stadium, $x$ and $y$ are such that:
$$
\left(\frac{2x}{\ell}\right)^n + \left(\frac{2y}{w}\right)^n = 1
$$

When $n$ get bigger, the shape will get flatter in the middle and at the tips of the cortex, note that $n = 2$ gives back the previous elliptical shape.

```python
def cortex2(length, width, exp=4., num_points=1000):
    if exp % 2 == 0:
        
        xs_up = np.linspace(-length/2., length/2., num_points // 2)
        ys_up = (width / 2.) * ((1 - (2. * xs_up/length)**exp)**(1./exp))
        ys = np.hstack((- ys_up[::-1], ys_up))
        xs = np.hstack((xs_up[::-1], xs_up))
    else:
        xs_up = np.linspace(0, length/2., num_points // 4)
        ys_up = (width/ 2.) * ((1 - (2. * xs_up/length)**exp)**(1./exp))
        ys = np.hstack((ys_up[::-1], ys_up,
                        - ys_up[::-1], -ys_up))
        xs = np.hstack((-xs_up[::-1], xs_up,
                        xs_up[::-1], -xs_up))
    
    return (xs, ys)
```

```python
fig, ax = plt.subplots()

xs, ys = cortex2(14, 4, exp=2.5)
ax.plot(xs, ys, 'b-', label=r'$n = 2.5$')

xs, ys = cortex2(14, 4, exp=3.)
ax.plot(xs, ys, 'r-', label=r'$n = 3$')
xs, ys = cortex2(14, 4, exp=4.)
ax.plot(xs, ys, 'k-', label=r'$n = 4$')
ax.legend()
ax.set_aspect('equal')
ax.set_ylim(-2.5, 5.);
```

## Answer 3: straight lines and circles.

Here we just say that the cortex is comprized of a linear part (the body) and two circle arcs. It's simple, but it's not analytical (it's not a single formula for the whole body)... It's also the most faithfull reproduction of the shape, and can be very usefull in a modeling context, or for 3D representations (as cylinders and spheres are usually already present in 3D drawing software as primitives).

```python
def cortex3(length, width, num_points=1000):
    x_body_up = np.linspace(0, (length - width)/2.,
                            num_points//8)
    y_body_up = np.ones_like(x_body_up) * width / 2.
    
    theta_tip = np.linspace(0, np.pi / 2, num_points//8)
    theta_tip = theta_tip[::-1]
    x_tip_up = width * np.cos(theta_tip)/2 + (length - width)/2.
    y_tip_up = width * np.sin(theta_tip)/2
    
    xs_up = np.hstack((x_body_up, x_tip_up))
    ys_up = np.hstack((y_body_up, y_tip_up))
    
    ys = np.hstack((ys_up[::-1], ys_up,
                    - ys_up[::-1], -ys_up))
    xs = np.hstack((-xs_up[::-1], xs_up,
                    xs_up[::-1], -xs_up))
    
    return (xs, ys)
```

```python
xs, ys = cortex3(14, 4)
fig, ax = plt.subplots()
ax.plot(xs, ys, 'k-')

ax.set_aspect('equal')
ax.set_ylim(-2.5, 2.5);
```

## A quick note about ellipses

One of the good points of the analytical representation is that it allows to easyly rotate the shape by changing a little bit the ellipsis definition (i.e. by adding a phase term). That can be usefull for fitting purposes, where we try to get the shape from a contour extracted from an image...

```python
def cortex1bis(length, width, phase=0, num_points=100):
    thetas = np.linspace(0, 2*np.pi, num_points)
    xs = length/2. * np.cos(thetas)
    ys = width/2. * np.sin(thetas + phase)
    return xs, ys
```

```python
xs, ys = cortex1bis(14, 4, phase = np.pi/6)
fig, ax = plt.subplots()
ax.plot(xs, ys, 'k-')

ax.set_aspect('equal')
ax.set_ylim(-2.5, 2.5);
```

## 3D representation...

All the above representation can be extrapolated to 3D just by adding a rotation around the $x$ axis 

```python
np.meshgrid?
```

```python
def rotate(xs, rs, n_seg=12):
    phis = np.linspace(0, 2*np.pi, n_seg)
    x_grid, phi_grid = np.meshgrid(xs, phis)
    r_grid, phi_grid =  np.meshgrid(rs, phis)
    
    y_grid = r_grid * np.cos(phi_grid)
    z_grid = r_grid * np.sin(phi_grid)
    return x_grid, y_grid, z_grid
    
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(3, 1,
                                    subplot_kw={'projection':'3d'}, 
                                    figsize=(6, 12))
length = 14
width = 4

xs, rs = cortex1(length, width, num_points=50)
xs, ys, zs = rotate(xs, rs)
ax0.plot_wireframe(xs, ys, zs,
                   alpha=0.4, color='k')
ax0.set_title('Ellipsis')
#ax0.set_aspect('equal')


xs, rs = cortex2(length, width, exp=3, num_points=50)
xs, ys, zs = rotate(xs, rs)
ax1.plot_wireframe(xs, ys, zs,
                   alpha=0.4, color='k')
ax1.set_title('Flatten ellipsis')
#ax1.set_aspect('equal')

xs, rs = cortex3(length, width, num_points=50)
xs, ys, zs = rotate(xs, rs)
ax2.plot_wireframe(xs, ys, zs,
                   alpha=0.4, color='k')
ax2.set_title('Sphere and cylinder')
#ax2.set_aspect('equal')

```
