```python
import numpy as np
import pandas as pd
```

```python
index = pd.MultiIndex.from_tuples([(i, j) for i in range(3) for j in range(4)], names=['u', 'v'])
df = pd.DataFrame(np.random.random((12, 3)), columns=['A', 'B', 'C'], index=index)
```

```python
df
```

```python
fname = 'test.xlsx'
df.to_excel(fname, 'Sheet1', merge_cells=True)

df_r = pd.read_excel(fname, has_index_names=True)
```
