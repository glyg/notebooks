```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np, scipy as sp
```

```python
data = pd.read_csv('chameleon_moving_avg1.csv', header=None)
data[6:].plot()
```

```python
tr = np.fft.fft(data[6:], axis=0)

# Enhance the long period by correcting the mean and taking abs value
tr_2 = np.fft.fft(np.abs(data[6:]-data[6:].mean()), axis=0)

# Get the corresponding frequencies (the x axis bellow)
freq = np.fft.fftfreq(data[6:].size)

fig, ax = plt.subplots()
ax.plot(freq[:freq.size//2], np.abs(tr[:freq.size//2]))
ax.plot(freq[:freq.size//2], np.abs(tr_2[:freq.size//2]))

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim(1e-4, 3e-2)
```
